```python
# code
import numpy as np
import matplotlib.pyplot as plt

filename = 'fakeData.txt'
data     = np.genfromtxt(filename, delimiter = ',', autostrip = True, skip_header = 1)

# masks for filtering
mask_bad  = data[:, 2] == -1
mask_good = data[:, 2] == 1

# filter data into bad and good
bad_data  = data[mask_bad]    # -1 in the last col.
good_data = data[mask_good]   # 1 in the last col.

x_bad  = bad_data[:, 0]
y_bad  = bad_data[:, 1]
x_good = good_data[:, 0]
y_good = good_data[:, 1]

plt.plot(x_bad, y_bad, 'x', label = 'bad data')
plt.plot(x_good, y_good, 'o', label ='good data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = 'upper left')
plt.show()
```
